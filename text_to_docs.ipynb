```python
import subprocess
import docx
```

```python
def convert_latex_to_word(latex_file, output_file):
    # Convert LaTeX to Word using Pandoc
    subprocess.run(['pandoc', latex_file, '-s', '-o', 'temp.docx'])
    #subprocess.run(['pandoc', '-o', latex_file, 'temp.docx'])

    # Open the temporary Word document
    doc = docx.Document('/content/temp.docx')

    # Save the Word document with the desired output filename
    doc.save(output_file)

    # Delete the temporary Word document
    subprocess.run(['rm', '/content/temp.docx'])
```

```python
# Example usage
latex_file = '/content/input.tex'
output_file = '/content/output.docx'

convert_latex_to_word(latex_file, output_file)
```

```python
def convert_TEX_to_DOC(path_infile, path_outfile):
  #path_infile = '/content/' + infile
  #path_outfile = '/content/' + outfile

  options = ap.TeXLoadOptions()

  # Open TeX document with created load options
  document = ap.Document(path_infile, options)

  option = ap.DocSaveOptions()
  option.Format = ap.DocSaveOptions.DocFormat.Doc

  # Save the file into MS Word document format

  document.Save(path_outfile, option)
  print(path_infile + " converted into " + path_outfile)
```

```python
latex_file = '/content/input.tex'
output_file = '/content/output.docx'

convert_TEX_to_DOC(latex_file, output_file)
```
